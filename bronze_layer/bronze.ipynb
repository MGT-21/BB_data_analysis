```python
import pandas as pd

df = pd.read_csv('../raw_data/databridge_squad19_sintetico.csv')
```

```python
df["doc_type"] = df["doc_type"].str.upper()
df["source_system"] = df["source_system"].str.upper()
df["text_language"] = df["text_language"].str.upper()
df["ocr_engine"] = df["ocr_engine"].str.upper().str.strip()
```

```python
df["file_size_kb"] = pd.to_numeric(df["file_size_kb"], errors="coerce")
```

```python
df["normalized_date"] = pd.to_datetime(df["normalized_date"], errors="coerce")
```

```python
df.to_csv("databridge_squad19_bronze.csv", index=False)
```
